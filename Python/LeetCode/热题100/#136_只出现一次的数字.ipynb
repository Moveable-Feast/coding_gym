# 解答

```python
class Solution:
    def singleNumber(self, nums: list[int]) -> int:
        hash = dict()
        for num in nums:
            if num not in hash:
                hash[num] = 1
            else:
                hash[num] += 1
        for num in hash:
            if hash[num] == 1:
                return num
```

# 官方

方法一：位运算

```python
from functools import reduce

class Solution:
    def singleNumber(self, nums: list[int]) -> int:
        return reduce(lambda x, y: x ^ y, nums)
```

# 测试

```python
solution = Solution()
solution.singleNumber([4, 1, 2, 1, 2])
```
