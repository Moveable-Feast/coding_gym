# 解答

```python
class MedianFinder:

    def __init__(self):
        self.stream = []

    def addNum(self, num: int) -> None:
        self.stream.append(num)

    def findMedian(self) -> float:
        stream_sorted = self.stream.copy()
        stream_sorted.sort()
        l = len(stream_sorted)
        if len(stream_sorted) % 2 == 1:
            return float(stream_sorted[l >> 1])
        else:
            return (stream_sorted[l >> 1] + stream_sorted[(l >> 1) - 1]) / 2

# Your MedianFinder object will be instantiated and called as such:
# obj = MedianFinder()
# obj.addNum(num)
# param_2 = obj.findMedian()
```
