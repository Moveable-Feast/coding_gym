# 解答

```python
class Solution:
    def orangesRotting(self, grid: list[list[int]]) -> int:
        # 在循环内判断停止，即所有2旁边还有没有1
        # 在循环外统计时间
        minute = 0
        m = len(grid)
        n = len(grid[0])
        # 轮数
        while 1:
            unsafe = 0
            # 统计下一轮即将腐烂个数
            for i in range(m):
                for j in range(n):
                    if grid[i][j] == 1 and ((i - 1 >= 0 and grid[i - 1][j] == 2) or (i + 1 < m and grid[i + 1][j] == 2) or (j - 1 >= 0 and grid[i][j - 1] == 2) or (j + 1 < n and grid[i][j + 1] == 2)):
                        unsafe += 1
            
            if unsafe == 0:
                break           
            else:
                # 腐烂继续
                for i in range(m):
                    for j in range(n):
                        if grid[i][j] == 2:
                            if i - 1 >= 0 and grid[i - 1][j] == 1:
                                grid[i - 1][j] = '2'
                            if i + 1 < m and grid[i + 1][j] == 1:
                                grid[i + 1][j] = '2'
                            if j - 1 >= 0 and grid[i][j - 1] == 1:
                                grid[i][j - 1] = '2'
                            if j + 1 < n and grid[i][j + 1] == 1:
                                grid[i][j + 1] = '2'
            # '2'换成2
            for i in range(m):
                for j in range(n):
                    if grid[i][j] == '2':
                        grid[i][j] = 2
            # 轮数
            minute += 1
        unrotten = 0
        # 判断有无未腐烂
        for i in range(m):
            for j in range(n):
                if grid[i][j] == 1:
                    unrotten += 1
        if unrotten == 0:
            return minute
        else:
            return -1
```

# 官方

方法一：多源广度优先搜索

```python
import collections

class Solution:
    def orangesRotting(self, grid: list[list[int]]) -> int:
        R, C = len(grid), len(grid[0])

        # queue - all starting cells with rotting oranges
        queue = collections.deque()
        for r, row in enumerate(grid):
            for c, val in enumerate(row):
                if val == 2:
                    queue.append((r, c, 0))

        def neighbors(r, c) -> (int, int):
            for nr, nc in ((r - 1, c), (r, c - 1), (r + 1, c), (r, c + 1)):
                if 0 <= nr < R and 0 <= nc < C:
                    yield nr, nc

        d = 0
        while queue:
            r, c, d = queue.popleft()
            for nr, nc in neighbors(r, c):
                if grid[nr][nc] == 1:
                    grid[nr][nc] = 2
                    queue.append((nr, nc, d + 1))

        if any(1 in row for row in grid):
            return -1
        return d
```
