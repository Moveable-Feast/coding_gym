# 解答

```python
class Solution:
    def maxSubArray(self, nums: list[int]) -> int:
        n = len(nums)
        ans = nums[0]
        for i in range(n):
            for j in range(i, n):
                ans = max(sum(nums[i : j + 1]), ans)
        return ans
```

# 官方

方法一：动态规划

```python
class Solution:
    def maxSubArray(self, nums: list[int]) -> int:
        pre = 0; maxAns = nums[0]
        for x in nums:
            pre = max(pre + x, x)
            maxAns = max(maxAns, pre)
        return maxAns
```
